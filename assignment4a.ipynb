```python
df.drop(['aa_000', 'ab_000','ad_000', 'bk_000', 'bm_000', 'bn_000', 'bo_000', 'bp_000', 'bq_000', 'br_000', 'cf_000', 'cg_000', 'co_000', 'cr_000', 'ct_000', 'cu_000', 'cv_000', 'cx_000', 'cz_000', 'da_000', 'db_000', 'dc_000'], axis=1, inplace=True)
print (df)
```

```python
df = df.replace('pos', 1)
df = df.replace('neg', 0)
```

```python
X = df.drop('class', axis=1)
y = df['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
##df = df.apply(pd.to_numeric, errors='coerce')
##df = df.fillna(df.mean())
X_train = X_train.apply(pd.to_numeric, errors='coerce')
X_train = X_train.fillna(X_train.mean())
X_test = X_test.apply(pd.to_numeric, errors='coerce')
X_test = X_test.fillna(X_train.mean())

y_train = y_train.apply(pd.to_numeric, errors='coerce')
y_train = y_train.fillna(X_train.mean())
y_test = y_test.apply(pd.to_numeric, errors='coerce')
y_test = y_test.fillna(X_train.mean())
```

```python
print (X_train, X_test, y_train, y_test)
```

```python
svclass = SVC(kernel='poly', degree=8)
svclass.fit(X_train, y_train)
```

```python
y_pred = svclass.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
